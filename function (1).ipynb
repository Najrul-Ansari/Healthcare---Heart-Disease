```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,precision_score, mean_squared_error

## importing data
def read_data():
    try:
        df = pd.read_csv(r"C:\Users\Pc\OneDrive\Pictures\Data science assignments\week 18\Assignments\heart.csv")
        return df
    except Exception as e:
        print("Data path Not Found and the error is in Read Data Function")

## Preparing the data for model
def data_prep(df, t):
    try:
        X = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
                 'exang', 'oldpeak', 'slope', 'ca', 'thal']]
        y = df['target']
        X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=t, random_state=2586)
        return X_train,X_test,y_train,y_test
    except Exception as e:
        print("The error is in Data Prep Function")

## Logistic Regression
def log_model(X_train,y_train,X_test):
    log_model = LogisticRegression(max_iter=1000, solver='sag', penalty='l2')
    log_model_fit = log_model.fit(X_train,y_train)
    pred = log_model_fit.predict(X_test)
    return pred

## Decision tree
def decision_tree(X_train,y_train,X_test):
    dt_model = DecisionTreeClassifier(criterion='entropy',max_depth =3)
    dt_model_fit = dt_model.fit(X_train,y_train) 
    pred = dt_model_fit.predict(X_test)
    return pred

## Random Forest
def rf_classifier(X_train,y_train,X_test):
    rf_model = RandomForestClassifier(criterion='entropy',max_depth =5)
    rf_model_fit = rf_model.fit(X_train,y_train)
    pred = rf_model_fit.predict(X_test)
    return pred

## KNeighborsClassifier
def kn_classifier(X_train,y_train,X_test):
    knc_model = KNeighborsClassifier(n_neighbors=15)
    knc_model_fit = knc_model.fit(X_train,y_train)
    pred = knc_model_fit.predict(X_test)
    return pred

## SVM
def svm_model(X_train,y_train,X_test):
    svm_model = svm.SVC(kernel='linear')
    svm_model_fit = svm_model.fit(X_train,y_train)
    pred = svm_model_fit.predict(X_test)
    return pred

## GaussianNB()
def naive_bayes_model(X_train,y_train,X_test):
    nb_model = GaussianNB()
    nb_model_fit = nb_model.fit(X_train, y_train)
    pred = nb_model_fit.predict(X_test)
    return pred

## Model Diagnostic
def model_diag(y_test,pred):
    acc = accuracy_score(y_test,pred)
    prc = precision_score(y_test,pred)
    clf = classification_report(y_test,pred)
    conf = confusion_matrix(y_test,pred)
    rmse = np.sqrt(mean_squared_error(y_test,pred))
    return acc, prc,clf,conf, rmse
```
